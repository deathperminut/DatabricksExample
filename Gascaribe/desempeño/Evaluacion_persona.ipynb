```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
df=pd.read_excel('Competencia.xls')
```

```python
df.head()
```

```python
def promedio_evaluado(df):
    df=df[['Evaluado','Rut','Nota']]
    df=df.groupby(['Evaluado','Rut'])['Nota'].mean().to_frame().reset_index()
    return df
```

```python
promedio_evaluado(df)
```

```python
promedio_evaluado(df).mean()
```

```python
sns.displot(data=promedio_evaluado(df),x='Nota',kind='kde')
```
